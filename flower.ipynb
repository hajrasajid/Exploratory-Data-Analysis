```python
#Box plot
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

plt.figure(figsize=(15,6))

for i, col in enumerate(cols):
    plt.subplot(1, 5, i+1)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
#Remove outliers
col = 'SepalWidthCm'

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_clean = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("Before shape:", df.shape)
print("After shape:", df_clean.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df_clean['SepalWidthCm'])
plt.title("After Removing Outliers")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

plt.figure(figsize=(20, 4))

colors = ["blue", "green", "orange", "red"]

for i, col in enumerate(num_cols):
    plt.subplot(1, 4, i+1)

    sns.histplot(
        df_clean[col],
        kde=True,
        bins=20,
        color=colors[i]
    )

    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Correlation heatmap
plt.figure(figsize=(6,5))
sns.heatmap(df_clean[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

species_list = df_clean["Species"].unique()

# Assign colors to each species
colors = ["blue", "green", "red"]

plt.figure(figsize=(15,4))

for i, sp in enumerate(species_list):
    plt.subplot(1, len(species_list), i+1)

    data_sp = df_clean[df_clean["Species"] == sp]

    sns.regplot(
        data=data_sp,
        x="SepalLengthCm",
        y="SepalWidthCm",
        scatter_kws={"alpha": 0.6, "color": colors[i]},
        line_kws={"color": colors[i]}
    )

    plt.title(sp)
    plt.xlabel("Sepal Length")
    plt.ylabel("Sepal Width")

plt.tight_layout()
plt.show()
```
